```python
import os
from crewai import Agent, Task, Crew, Process
from crewai_tools import SerperDevTool
```

```python
search_tool= SerperDevTool()
```

```python
from langchain_openai import ChatOpenAI
```

```python
researcher = Agent(
    role="Research Assistant",
    goal="Look up advancements in AI agents",
    backstory="""You work with a professor of AI at Stanford University. You have knowledge about NLP and Deep Learning and have already completed several projects related to the same.""",
    verbose=True,
    allow_delegations=False,
    tools=[search_tool],
    llm=ChatOpenAI(model_name="gpt-4o", temperature=0.3),
)
```

```python
writer= Agent(
    role="Professional Article Writer",
    goal="Crafting compelling content on tech advancements",
    backstory="""You are a known tech writer, who writes about new advancements in the field of AI, and you have published more than 100 articles on medium.""",
    verbose=True,
    allow_delegations=True,
    llm=ChatOpenAI(model_name="gpt-4o", temperature=0.7)
)
```

```python
task1= Task(
    description="""Conduct a research about agents used for automating github.""",
    expected_output="Full analysis for already present agents that automate github repository creation and managing commits.",
    agent=researcher
)
```

```python
task2= Task(
    description="""Using the insights provided, write an article about how an agent can be created about creating github repositories and push and commit our code on github.""",
    expected_output="Full blog post of atleast 5 paragraphs",
    agent= writer
)
```

```python
crew= Crew(
    agents=[researcher, writer],
    tasks= [task1, task2],
    verbose=1
)
```

```python
result= crew.kickoff()
print("################")
print(result)
```
